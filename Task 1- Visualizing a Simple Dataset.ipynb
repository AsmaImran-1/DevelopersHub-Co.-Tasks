```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Loading the Iris dataset from seaborn
iris_df = sns.load_dataset('iris')
```

```python
print("--- DataFrame Shape ---")
print(iris_df.shape)

print("\n--- Column Names ---")
print(iris_df.columns)

print("\n--- First 5 Rows (Head) ---")
print(iris_df.head())
```

```python
# SCATTER PLOT (Relationship)
Objective: Analyzing the relationship between petal length and petal width.
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x='petal_length', 
    y='petal_width', 
    data=iris_df, 
    hue='species', 
    palette='viridis', 
    s=100
)
plt.title('Scatter Plot: Petal Length vs. Petal Width by Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species')
plt.grid(True)
plt.show()
```

```python
# HISTOGRAM (Data Distribution)
Objective: Examine the distribution of sepal length.
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(
    data=iris_df, 
    x='sepal_length', 
    kde=True, # Add a kernel density estimate line
    bins=15, 
    color='skyblue'
)
plt.title('Histogram of Sepal Length Distribution')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# BOX PLOT (Outliners and Spread)
Objective: Compare the distribution and detect outliers in sepal width across the three species.
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(
    x='species', 
    y='sepal_width', 
    data=iris_df, 
    hue='species', 
    palette='Set2',
    legend=False 
)
plt.title('Box Plot of Sepal Width by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Width (cm)')
plt.show()
```
